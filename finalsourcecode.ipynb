```python
import gradio as gr
import pickle
with open ("model_pickle","wb") as f:
    pickle.dump(m,f)
with open ("model_pickle","rb") as f:
    z=pickle.load(f)
```

```python
def pred(salary,saving,expenditure,paid,accountbalance,Credit_score):
    x=np.array([salary+saving+expenditure+paid+accountbalance])
    pred=z.predict([x])
    while (pred==1 and Credit_score >=6 and x>=200000 ) :
        return "probalblity is high"
    else:
        return "probalblity is less"
output=gr.outputs.Textbox()
app=gr.Interface(fn=pred,inputs=['number','number','number','number','number','number'],outputs=output)
app.launch()
```
